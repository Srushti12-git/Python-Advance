1. What is NumPy, and why is it widely used in Python?

Ans:-NumPy (Numerical Python) is a powerful open-source library in Python mainly used for numerical computations.
Why is NumPy widely used?
✅ Efficient Array Operations:

✅ Memory Efficient:
NumPy arrays use less memory than Python lists for storing data.

✅ Foundation for Other Libraries:
Used as the base for libraries like Pandas, SciPy, TensorFlow, and scikit-learn.

✅ Interfacing with C/C++/Fortran:
Offers tools to integrate with low-level languages for high-performance computing.

2. How does broadcasting work in NumPy?

Ans:-Broadcasting is a powerful feature that allows NumPy to perform operations on arrays of different shapes as if they had the same shape, without copying data.
✅ How It Works:
When performing element-wise operations:

a.  NumPy compares array shapes from right to left.

b.  The dimensions are compatible if:

=>They are equal, or

=>One of them is 1.

c.  If compatible, NumPy stretches (broadcasts) the smaller array along the mismatched dimension.

6. What are the differences between NumPy arrays and Python lists?

Ans:-
| Feature                      | **NumPy Array**                                                       | **Python List**                                        |
| ---------------------------- | --------------------------------------------------------------------- | ------------------------------------------------------ |
| **Data Type**                | Homogeneous (all elements of same type)                               | Heterogeneous (can store mixed types)                  |
| **Performance**              | Faster (uses C-optimized libraries)                                   | Slower (pure Python objects)                           |
| **Memory Usage**             | Less memory (compact data storage)                                    | More memory (stores object references)                 |
| **Vectorized Operations**    | Supports element-wise operations directly                             | Needs explicit loops or list comprehensions            |
| **Multidimensional Support** | Built-in (e.g., matrices, tensors)                                    | Supports only nested lists (no true multi-dim support) |
| **Broadcasting**             | Yes (automatic resizing for operations)                               | No broadcasting feature                                |
| **Built-in Functions**       | Rich mathematical/statistical operations (`sum()`, `mean()`, `dot()`) | Only basic list methods                                |

18. What are some advanced features of NumPy?

Ans:-
✅ 1. Broadcasting
=>Perform operations on arrays of different shapes without copying data.

=>Enables vectorized computations on mismatched shapes.

a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)  # Broadcasted to (3, 3)

✅ 2. Advanced Indexing & Boolean Masking
Select or modify array elements using conditions or index arrays.

a = np.array([1, 2, 3, 4])
print(a[a > 2])  # [3 4]

✅ 3. Linear Algebra Functions (numpy.linalg)
Matrix multiplication, eigenvalues, determinants, inverse.

from numpy.linalg import inv
a = np.array([[1, 2], [3, 4]])
print(inv(a))

✅ 4. Random Sampling (numpy.random)
Generate random numbers, distributions, sampling.

np.random.seed(0)
print(np.random.normal(0, 1, size=5))

✅ 5. Fourier Transforms (numpy.fft)
For frequency domain analysis.

np.fft.fft([0, 1, 0, 0])

✅ 6. Memory Mapping of Large Files (np.memmap)
Read large binary data without loading it all into RAM.

✅ 7. Vectorized Universal Functions (ufuncs)
Optimized functions that operate element-wise.

np.add(a, a)
np.exp(a)

✅ 8. Structured Arrays & Record Arrays
Arrays with heterogeneous types (like database records).

dt = np.dtype([('name', 'U10'), ('age', 'i4')])
data = np.array([('Alice', 25), ('Bob', 30)], dtype=dt)

21. Why is NumPy’s array slicing faster than Python’s list slicing?

Ans:-
1.  Contiguous Memory Allocation
=>NumPy arrays are stored in contiguous blocks of memory, making them highly cache-efficient.
=>Python lists are arrays of pointers to objects scattered in memory, causing slower access and slicing.

2.  Fixed Data Types (Homogeneous Elements)
=>NumPy arrays hold elements of the same data type, allowing operations like slicing to work on raw memory buffers.
=>Python lists are heterogeneous, so slicing involves copying references to individual objects.

3.  View vs Copy
=>NumPy slicing returns a view (a lightweight object referencing the same data), so it doesn’t copy the data.
=>Python list slicing always creates a new list with copied elements, which takes more time and memory.

4.  Low-Level Optimizations
=>NumPy is implemented in C, using optimized algorithms with vectorized operations.
=>Python list operations are implemented in high-level Python, which adds interpretation overhead
